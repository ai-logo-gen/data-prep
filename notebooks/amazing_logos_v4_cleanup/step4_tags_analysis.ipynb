```python
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter
import ast

# Define paths
metadata_path = Path('../../output/amazing_logos_v4/data/amazing_logos_v4_cleanup/metadata8.csv')

print(f"Checking if metadata file exists: {metadata_path}")
print(f"File exists: {metadata_path.exists()}")

if metadata_path.exists():
    print(f"File size: {metadata_path.stat().st_size / 1024 / 1024:.2f} MB")
else:
    print("File not found! Please check the path.")
```

```python
# Load the metadata
print("Loading metadata7...")
df = pd.read_csv(metadata_path)

print(f"Dataset shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nFirst few rows:")
print(df.head())

# Check for tags and categories columns
tags_cols = [col for col in df.columns if 'tag' in col.lower()]
category_cols = [col for col in df.columns if 'categor' in col.lower()]

print(f"\nColumns containing 'tag': {tags_cols}")
print(f"Columns containing 'categor': {category_cols}")

# Show basic info about these columns
for col in tags_cols + category_cols:
    if col in df.columns:
        print(f"\n{col}:")
        print(f"  Non-null count: {df[col].notna().sum()}")
        print(f"  Sample values: {df[col].dropna().head(3).tolist()}")
```

```python
# Extract unique categories
print("Extracting unique categories...")
unique_categories = set(df['category'].unique())
print(f"Number of unique categories: {len(unique_categories)}")
print(f"Categories: {sorted(list(unique_categories))}")

print("\n" + "="*80)

# Extract unique tags from the comma-separated tags column
print("Extracting unique tags...")
all_tags = []

for tags_str in df['tags'].dropna():
    if pd.notna(tags_str) and isinstance(tags_str, str):
        # Split by comma and clean up whitespace
        tags = [tag.strip() for tag in tags_str.split(',') if tag.strip()]
        all_tags.extend(tags)

# Count tag occurrences
tag_counts = Counter(all_tags)
unique_tags = set(all_tags)

print(f"Number of unique tags: {len(unique_tags)}")
print(f"Total tag occurrences: {len(all_tags)}")
print(f"\nMost common tags:")
for tag, count in tag_counts.most_common(20):
    print(f"  {tag}: {count:,}")

print("\n" + "="*80)
```

```python
# Find tags that are NOT present in the categories list
print("Analyzing tags not present in categories...")

tags_not_in_categories = unique_tags - unique_categories

print(f"Number of tags not in categories: {len(tags_not_in_categories)}")
print(f"Number of tags that ARE in categories: {len(unique_tags & unique_categories)}")

# Create a list of tags not in categories with their counts, sorted by count
tags_not_in_categories_with_counts = [
    (tag, tag_counts[tag]) for tag in tags_not_in_categories
]

# Sort by count (descending)
tags_not_in_categories_with_counts.sort(key=lambda x: x[1], reverse=True)

print(f"\nTags NOT present in categories list (sorted by appearance count):")
print("=" * 80)
print(f"{'Tag':<40} {'Count':<10}")
print("-" * 80)

for tag, count in tags_not_in_categories_with_counts:
    print(f"{tag:<40} {count:,}")

# Also show some statistics
total_occurrences_not_in_categories = sum(count for _, count in tags_not_in_categories_with_counts)
print(f"\nStatistics:")
print(f"Total occurrences of tags not in categories: {total_occurrences_not_in_categories:,}")
print(f"Percentage of tag occurrences not in categories: {(total_occurrences_not_in_categories/len(all_tags)*100):.2f}%")
```
